##Solution

###a)

```python
from IPython.core.display import Image 
Image(filename='4.png') 
```

in x:

$$P_{x}=l_{1}cos(\alpha_{1})-l_{2}cos(180°-(\alpha_{2}+\alpha_{1}))$$
 
$$P_{x}=l_{1}cos(\alpha_{1})+l_{2}cos(-\alpha_{2}-\alpha_{1}) = l_{1}cos(\alpha_{1})+l_{2}cos(\alpha_{2}+\alpha_{1})$$

in y:  
$$P_{y}=-l_{1}sin(\alpha_{1})-l_{2}sin(180°-(\alpha_{2}+\alpha_{1}))$$

$$P_{y}=-l_{1}sin(\alpha_{1})+l_{2}sin(-\alpha_{2}-\alpha_{1}) = -l_{1}sin(\alpha_{1})-l_{2}sin(\alpha_{2}+\alpha_{1})$$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from IPython.display import display, Math
from sympy import Symbol, symbols, Function, Matrix, simplify, lambdify, expand, latex
from sympy import diff, cos, sin, sqrt, acos, atan2, atan
from sympy.physics.mechanics import dynamicsymbols, mlatex

t = Symbol('t')
l1, l2 = symbols('ell_1 ell_2', positive=True)
a1, a2 = dynamicsymbols('alpha1 alpha2')  # alpha1(t) alpha2(t)
```

```python
p = Matrix((l1*cos(a1)+l2*cos(a1+a2),-l1*sin(a1)-l2*sin(a1+a2)))
Math(mlatex(r'\mathbf{p}=\begin{bmatrix} P_x \\ P_y \end{bmatrix}=') + mlatex(p))
```

###b)

$$ \mathbf{J}= 
\begin{bmatrix}
\frac{\partial \:P_x}{\partial \alpha_1}\frac{\partial \:P_x}{\partial \alpha_2} \\
\frac{\partial \:P_y}{\partial \alpha_1}\frac{\partial \:P_y}{\partial \alpha_2} \\
\end{bmatrix} $$

```python
J = p.jacobian([a1, a2])
Math(mlatex(r'\mathbf{J}=') + mlatex(J))
```
